## Коротко о теории

### 1. Логистическая регрессия
Логистическая регрессия моделирует вероятность положительного класса:

$
P(y=1 \mid x) = \sigma(z), \quad z = w^T x + b
$

где \(\sigma(z)\) — сигмоида:

$
\sigma(z) = \frac{1}{1 + e^{-z}}
$

### 2. Почему не линейная регрессия
Если предсказывать класс линейной моделью напрямую, прогноз может выйти за диапазон \([0, 1]\).  
Сигмоида решает эту проблему: она переводит любое число в вероятность.

### 3. Функция потерь
Для логистической регрессии используется **log-loss**:

$
L = - \frac{1}{n} \sum_{i=1}^{n}
\left[
y_i \log(\hat{p}_i) + (1-y_i)\log(1-\hat{p}_i)
\right]
$

### 4. Обучение
Мы будем минимизировать log-loss методом **градиентного спуска**.

### 5. Что будем сравнивать
После своей реализации вы сравните результат с `sklearn.linear_model.LogisticRegression`:
- accuracy,

- ROC-AUC.

### 6. Почему нужна стандартизация
Градиентный спуск работает стабильнее, когда признаки находятся примерно в одном масштабе.

## Шаг 2. Реализуем сигмоиду

Допишите функцию `sigmoid(z)`.

Проверка:
- `sigmoid(0)` должно быть равно `0.5`;
- функция должна работать и со скалярами, и с `numpy`-массивами.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

print("sigmoid(0) =", sigmoid(0))
print("sigmoid([-1, 0, 1]) =", sigmoid(np.array([-1.0, 0.0, 1.0])))
```

## Шаг 3. Реализуем log-loss

Напишите функцию `compute_log_loss(y_true, y_pred_proba)`.

Важно:
- чтобы избежать `log(0)`, используйте `np.clip(y_pred_proba, 1e-15, 1 - 1e-15)`.

```python
def compute_log_loss(y_true, y_pred_proba):
    y_pred_proba = np.clip(y_pred_proba, 1e-15, 1 - 1e-15)
    loss = -(y_true * np.log(y_pred_proba) + (1 - y_true) * np.log(1 - y_pred_proba))
    return np.mean(loss)

toy_y = np.array([0, 1, 1, 0])
toy_p = np.array([0.1, 0.9, 0.8, 0.3])
print("Toy log-loss:", compute_log_loss(toy_y, toy_p))
```

## Шаг 4. Один шаг градиентного спуска

Напомним:
- \(z = Xw + b\)
- \(\hat{p} = \sigma(z)\)

Градиенты для log-loss:
\[
dw = 
rac{1}{n} X^T(\hat{p} - y), \qquad db = 
rac{1}{n}\sum(\hat{p} - y)
\]

Ниже нужно реализовать функцию, которая считает:
- вероятности,
- loss,
- градиенты `dw`, `db`.

```python
def forward_backward(X, y, w, b):
    # YOUR CODE HERE
    # 1) посчитать z
    # 2) посчитать p = sigmoid(z)
    # 3) посчитать loss
    # 4) посчитать dw и db
    n = X.shape[0]
    z = X.dot(w) + b
    p = sigmoid(z)
    loss = compute_log_loss(y, p)
    dw = (1 / n ) * X.T @ (p - y)
    db = np.mean(p - y)
    return p, loss, dw, db

n_features = X_train_scaled.shape[1]
w0 = np.zeros(n_features)
b0 = 0.0

p0, loss0, dw0, db0 = forward_backward(X_train_scaled, y_train.values, w0, b0)
print("Initial loss:", loss0)
print("dw shape:", dw0.shape)
print("db:", db0)
```

## Шаг 5. Собираем модель в класс

Допишите методы:
- `fit`
- `predict_proba`
- `predict`

Требования:
- обучение через градиентный спуск;
- сохранять историю значения loss;
- в `predict` использовать порог `threshold`.

```python
class MyLogisticRegressionGD:
    def __init__(self, learning_rate=0.05, n_iters=3000, threshold=0.5):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.threshold = threshold
        self.w = None
        self.b = None
        self.loss_history = []

    def fit(self, X, y, lambda_reg = 0.1):
        # YOUR CODE HERE
        # 1) инициализируйте w и b
        # 2) в цикле вызовите forward_backward
        # 3) обновите параметры
        # 4) сохраните loss в self.loss_history
        n_features = X.shape[1]
        self.w = np.zeros(n_features)
        self.b = 0.0
        self.loss_history = []
        n = X.shape[0]

        for i in range (self.n_iters):
            p, loss, dw, db = forward_backward(X, y, self.w, self.b)
            dw_reg = dw + (lambda_reg / n) * self.w
            self.w -= self.learning_rate * dw_reg
            self.b -= self.learning_rate * db
            self.loss_history.append(loss + (lambda_reg / (2 * n)) * np.sum(self.w**2))
        return self

    def predict_proba(self, X):
        z = X @ self.w + self.b
        proba = sigmoid(z)

        return proba

    def predict(self, X):
        proba = self.predict_proba(X)
        preds = (proba >= self.threshold).astype(int)
        return preds
```

# Ответы
1. Почему сигмоида?
Она сжимает линейный ответ (w^\top x + b) в диапазон от 0 до 1, поэтому выход можно трактовать как вероятность класса 1 и согласовать с log-loss.

2. Почему log-loss, а не MSE?
Log-loss — стандартная функция для вероятностной постановки (бернуллиевская модель); она лучше наказывает уверенные ошибки. MSE для вероятностей не является типичным выбором и хуже отражает задачу бинарной классификации.

3. Зачем стандартизация перед градиентным спуском?
Признаки в разных шкалах дают разный масштаб градиентов; после стандартизации спуск обычно сходится быстрее и ведёт себя предсказуемее.

4. Насколько близки метрики к sklearn?
Результаты получились практиески идентичны. Данная модель показала даже немного лучше результат ROC AUC, но в остальном незначительно но лучше scikit-learn.

5. Почему библиотечная модель стабильнее и быстрее?
В sklearn стоят эффективные солверы (не чистый Python-GD), меньше итераций и лучше обработка плохо обусловленных задач; плюс по умолчанию часто есть L2, что добавляет устойчивость весам.

6. Как порог влияет на precision и recall?
Порог ниже — чаще предсказываем класс 1 → обычно растёт recall, падает precision. Порог выше — реже класс 1 → чаще растёт precision, падает recall.

7. Где логрег сильна, где слаба?
Сильна при почти линейной границе, когда нужны скорость и интерпретация весов. Слабее там, где нужны нелинейности, сложные взаимодействия, картинки/сырой текст без удачных признаков — там чаще выигрывают ансамбли / нейросети.

**Задание A.** В функцию потерь добавлен штраф L2: к градиенту по весам прибавляется `(λ/n)·w`, к истории loss — малый штраф `(λ/(2n))·‖w‖²`, что уменьшает переобучение и стабилизирует веса.

```python
class MyLogisticRegressionGD:
    def __init__(self, learning_rate=0.05, n_iters=3000, threshold=0.5):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.threshold = threshold
        self.w = None
        self.b = None
        self.loss_history = []

    def fit(self, X, y, lambda_reg=0.1):
        """lambda_reg — коэффициент L2 (как в задании). y — numpy-массив 0/1."""
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)
        self.b = 0.0
        self.loss_history = []

        y = np.asarray(y).ravel()

        for _ in range(self.n_iters):
            p, loss, dw, db = forward_backward(X, y, self.w, self.b)
            
            dw_reg = dw + (lambda_reg / n_samples) * self.w
            self.w -= self.learning_rate * dw_reg
            self.b -= self.learning_rate * db
           
            l2_term = (lambda_reg / (2 * n_samples)) * np.sum(self.w ** 2)
            self.loss_history.append(loss + l2_term)

        return self

    def predict_proba(self, X):
        return sigmoid(X @ self.w + self.b)

    def predict(self, X):
        return (self.predict_proba(X) >= self.threshold).astype(int)
```

```python
my_model = MyLogisticRegressionGD(learning_rate=0.01, n_iters=3000, threshold=0.5)
my_model.fit(X_train_scaled, y_train.values, lambda_reg=0.1)
```

### Задание C
Попробуйте:
- уменьшить число итераций;
- изменить `learning_rate`;
- посмотреть, как это влияет на сходимость и метрики.

```python
def eval_model(n_iters, lr, lambda_reg=0.1):
    m = MyLogisticRegressionGD(learning_rate=lr, n_iters=n_iters, threshold=0.5)
    m.fit(X_train_scaled, y_train.values, lambda_reg=lambda_reg)
    pred = m.predict(X_test_scaled)
    proba = m.predict_proba(X_test_scaled)
    return {
        "n_iters": n_iters,
        "lr": lr,
        "accuracy": accuracy_score(y_test, pred),
        "f1": f1_score(y_test, pred),
        "roc_auc": roc_auc_score(y_test, proba),
        "final_loss": m.loss_history[-1],
        "model": m,
    }


experiments = []
for n_iters in [200, 3000]:
    for lr in [0.001, 0.01, 0.3]:
        experiments.append(eval_model(n_iters, lr))

exp_df = pd.DataFrame([{k: v for k, v in e.items() if k != "model"} for e in experiments])
display(exp_df.round(4))


bad = experiments[0]["model"]
good = experiments[-1]["model"]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(bad.loss_history, label="мало итераций / неудачный lr")
ax.plot(good.loss_history, label="больше итераций / удачный lr")
ax.set_xlabel("Итерация")
ax.set_ylabel("Loss")
ax.legend()
plt.title("Задание C: сходимость")
plt.show()
```

**Задание C.** При **малом числе итераций** loss не успевает упасть — метрики на test ниже. При **слишком малом learning rate** кривая loss сходится долго. При **слишком большом** loss может осциллировать или расти — качество падает. Сравниваем таблицу метрик и форму `loss_history`.
